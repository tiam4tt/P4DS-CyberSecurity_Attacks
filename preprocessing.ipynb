```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('cybersecurity_attacks.csv')

print(df.shape)
df.info()
```

```python
df.describe()
```

```python
df.describe(include='object')
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
df['Year'] = df['Timestamp'].dt.year
df['Month'] = df['Timestamp'].dt.month
df['Day'] = df['Timestamp'].dt.day
df['Hour'] = df['Timestamp'].dt.strftime('%H:%M:%S')
df = df.drop(columns=['Timestamp'])

print(df.shape)
df.info()
```

```python
# Count duplicate rows
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)
# Identify columns with null values
columns_with_null = df.columns[df.isnull().any()].tolist()
```

```python
df['Alerts/Warnings'] = df['Alerts/Warnings'].apply(lambda x: 'unknown' if pd.isna(x) else x)
df['Malware Indicators'] = df['Malware Indicators'].apply(lambda x: 'unknown' if pd.isna(x) else x)
df['Proxy Information'] = df['Proxy Information'].apply(lambda x: 'unknown' if pd.isna(x) else x)
df['Firewall Logs'] = df['Firewall Logs'].apply(lambda x: 'unknown' if pd.isna(x) else x)
df['IDS/IPS Alerts'] = df['IDS/IPS Alerts'].apply(lambda x: 'unknown' if pd.isna(x) else x)

print(df.isnull().sum())
```
